```python
from nltk.stem import WordNetLemmatizer
lemm=WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
from nltk import pos_tag
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADJ
    else:
        return wordnet.NOUN
```

```python
from nltk.corpus import stopwords
import string
stops=set(stopwords.words('english'))
punctuations=list(string.punctuation)
stops.update(punctuations)
stops
```

```python
def clean_review(words):
    opp=[]
    for w in words:
        if w.lower() not in stops:
            pos=pos_tag([w])
            clean_word=lemm.lemmatize(w,pos=get_simple_pos(pos[0][1]))
            opp.append(clean_word.lower())
    return opp
```

```python
documents=[(clean_review(document),category) for document,category in documents]
```

```python
training_documents=documents[:1500]
testing_documents=documents[1500:]
```

```python
all_words=[]
for doc in training_documents:
    all_words+=doc[0]
    
```

```python
import nltk
freq=nltk.FreqDist(all_words)
len(freq.most_common())
common=freq.most_common(3000)
features=[i[0] for i in common]
features
```

```python
def get_feature_dic(words):
    current_feature={}
    words_set=set(words)
    for w in features:
        current_feature[w]=w in words_set
    return current_feature
```

```python
output=get_feature_dic(training_documents[0][0])
output
```

```python
training_data=[(get_feature_dic(doc),category) for doc, category in training_documents ]
testing_data=[(get_feature_dic(doc),category) for doc, category in training_documents ]
```

```python
from nltk import NaiveBayesClassifier
```

```python
classifier=NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(classifier,testing_data)
```
