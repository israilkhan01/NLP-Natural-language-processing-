```python
import numpy as np
```

```python
import pandas as pd
```

```python
xy_train=pd.read_csv('twitter_x_y_train.csv')
# xy_train
x_test=pd.read_csv('twitter_x_test.csv')
```

```python
x=x_test.text
x
```

```python
x_train=xy_train.text
y_train=xy_train.airline_sentiment
```

```python
x_train.shape[0]
```

```python
X_train=[]
for index in range(x_train.shape[0]):
    X_train.append((nltk.word_tokenize(x_train[index]),y_train[index]))
# X_train
```

```python
X_test=[]
for tx in x:
    X_test.append(nltk.word_tokenize(tx))
# X_test
```

```python
len(X_test)
```

```python
from nltk.stem import WordNetLemmatizer
lemm=WordNetLemmatizer()
```

```python
from nltk.corpus import stopwords
import string
stops=set(stopwords.words('english'))
punctuations=list(string.punctuation)
stops.update(punctuations)
# stops
```

```python
from nltk.corpus import wordnet
from nltk import pos_tag
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADJ
    else:
        return wordnet.NOUN
```

```python
def clean_review(words):
    opp=[]
    for w in words:
        if w.lower() not in stops:
            pos=pos_tag([w])
            clean_word=lemm.lemmatize(w,pos=get_simple_pos(pos[0][1]))
            opp.append(clean_word.lower())
    return opp
```

```python
Training_documents=[(clean_review(document),category) for document,category in X_train]
# Testing_documents=[(clean_review(document),category) for document,category in X_test]
```

```python
Training_documents[0]
```

```python
Testing_documents=[(clean_review(document)) for document in X_test]
```

```python
Testing_documents[0]
```

```python
all_words=[]
for doc in Training_documents:
    all_words+=doc[0]
    
```

```python
import nltk
freq=nltk.FreqDist(all_words)
len(freq.most_common())
common=freq.most_common(3000)
features=[i[0] for i in common]
features
```

```python
def get_feature_dic(words):
    current_feature={}
    words_set=set(words)
    for w in features:
        current_feature[w]=w in words_set
    return current_feature
```

```python
output=get_feature_dic(Training_documents[0][0])
output
```

```python
training_data=[(get_feature_dic(doc),category) for doc, category in Training_documents]
testing_data=[(get_feature_dic(doc)) for doc in Testing_documents ]
```

```python
from nltk import NaiveBayesClassifier
```

```python
classifier=NaiveBayesClassifier.train(training_data)
nltk.classify.accuracy(classifier,testing_data)
classifier.show_most_informative_features(15)
```

```python
from sklearn.svm import SVC
from nltk.classify.scikitlearn import SklearnClassifier
```

```python
svc=SVC(C=2,random_state=True,max_iter=250)
classifier_sk=SklearnClassifier(svc)
classifier_sk.train(training_data)
# nltk.classify.accuracy(classifier_sk,testing_data)
```

```python
predictions=classifier_sk.classify_many(testing_data)  
```
